## Uploading the Image

```python
from google.colab import files

print("Please upload the color image you want to process:")
uploaded = files.upload()

image_filename = next(iter(uploaded))
print(f"\nImage '{image_filename}' uploaded successfully.")
```

## Loading the Image and Converting to Grayscale

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the Image
original_image = cv2.imread(image_filename)

# Convert to Grayscale
gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

# Binarize the grayscale image using a fixed threshold
THRESHOLD_VALUE = 127
_, binary_image = cv2.threshold(gray_image, THRESHOLD_VALUE, 255, cv2.THRESH_BINARY)

# Display all three images
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# 1: Original Image
axes[0].imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original Image")
axes[0].axis("off")

# 2: Grayscale Image
axes[1].imshow(gray_image, cmap="gray")
axes[1].set_title("Grayscale Image")
axes[1].axis("off")

# 3: Binary Image
axes[2].imshow(binary_image, cmap="gray")
axes[2].set_title("Binary Image")
axes[2].axis("off")

plt.tight_layout()
plt.show()
```
